```python
string = "hellO my name is snjasbfj and i am a skfnskfsk"

v = ["a", "e", "i", "o", "u"]

counter = 0
for s in string.lower():
    if s in v:
        counter += 1

print(counter)
```

```python
bal = 100

while bal != 0:
    if bal < 0:
        print("you are broke")
        break
    x = int(input("Enetr how much you want: "))
    bal -= x 
    print(f"withdrawal done, now you have {bal}")
```

```python
to_find = 11

vals = [0, 1]
for i, val in enumerate(range(to_find+1)):
    if val < 2:
        continue
    vals.append(0)
    vals[i] = vals[i-2] + vals[i-1]

for i in vals:
    print(i)
```

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

d = {}

for i, j in zip(x, y):
    d[i] = j

for k, v in d.items():
    print(f"{k} -> {v}")
```
