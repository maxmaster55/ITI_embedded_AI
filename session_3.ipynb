```python
strs = ["test", "tests2", "dksdksjdksjdk"]

def get_longest_idx(strs):
    m = len(strs[0])
    for i in range(len(strs)):
        if len(strs[i]) > m:
            m = i
    
    return m


def get_longest_str(strs):
    m = strs[0]
    for i in strs:
        if len(i) > len(m):
            m = i
    
    return m


print(f"longest index is: {get_longest_idx(strs)}")
print(f"longest str is: {get_longest_str(strs)}")
```

```python
nums = [1, 2, 3, 4, 5, 11, 12]


def get_primes(nums):
    primes = []
    for i in nums:
        for j in range(2, i):
            if i % j == 0:
                break
        else:
            primes.append(i)
    return primes

get_primes(nums)
```

```python
vals = "0sssssssss00111222555"


def get_freq(val):
    freq = {}
    for i in val:
        if i in freq:
            freq[i] += 1
        else:
            freq[i] = 1
    return freq


get_freq(vals)
```

```python
class Product:
    name=""
    price=0
    quantity=0

    def __init__(self, name, quantity, price):
        self.name = name
        self.quantity= quantity
        self.price = price
    
    def restock(self, num):
        self.quantity += num
    
    def display(self):
        print(f"{self.name} -> num: {self.quantity}, price: {self.price}")
    def sell(self):
        if self.quantity > 0:
            self.quantity -= 1
        else: 
            print("nothing to sell")
    
    
```

```python
p1 = Product("carrots", 2, 10)
p1.sell()
p1.sell()
p1.display()
p1.sell()
p1.restock(10)
p1.sell()
p1.display()
```

```python
class Calculator:
    a = 0
    b = 0

    def __init__(self, a, b):
        self.a = a
        self.b = b
    
    def add(self):
        return self.a + self.b
    
    def sub(self):
        return self.a - self.b
    
    def div(self):
        if self.b == 0:
            print("error")
            return
        
        return self.a / self.b
    
    def mult(self):
        return self.a * self.b
    
    def display(self):
        print(f"{self.a} : {self.b}")
    
```

```python
c = Calculator(5, 6)

c.add()
```

```python
class Student:
    id = 0
    name = ""
    courses = []

    def __init__(self, id, name):
        self.id = id
        self.name = name
    
    def enroll(self, course):
        self.courses.append(course)
        
    def display(self):
        print(f"{self.id} -> name: {self.name}, courses: {self.courses}")

    
        
```

```python
class Bank:

    def __init__(self, name, address, balance):
        self.name = name
        self.balance = balance
        self.address = address
        pass

    def deposit(self, how_much):
        self.balance = how_much

    def withdraw(self, amount):
        if self.balance > amount:
            self.balance -= amount
```
