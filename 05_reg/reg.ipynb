```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import  mean_absolute_error, root_mean_squared_error, mean_squared_error
```

```python
np.random.seed(42)
x = np.random.rand(100,1)*10
x
```

```python
y_act = 2.5*(x**2) + 0.5*x + np.random.randn(100, 1)
y_act
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y_act,test_size=0.2,random_state=42)
```

```python
poly_feat = PolynomialFeatures(degree=100)
x_train_poly = poly_feat.fit_transform(x_train)
x_test_poly = poly_feat.fit_transform(x_test)
```

```python
model = LinearRegression()
model.fit(x_train_poly, y_train)
```

```python
y_train_pred = model.predict(x_train_poly)
y_test_pred = model.predict(x_test_poly)
```

```python
print(root_mean_squared_error(y_train, y_train_pred))
print(root_mean_squared_error(y_test, y_test_pred))
```

```python
print(mean_absolute_error(y_train, y_train_pred))
print(mean_absolute_error(y_test, y_test_pred))
```

```python
print(mean_squared_error(y_train, y_train_pred))
print(mean_squared_error(y_test, y_test_pred))
```

```python
lasso = Lasso(alpha=0.1)
lasso.fit(x_train_poly, y_train)

y_train_pred = lasso.predict(x_train_poly)
y_test_pred = lasso.predict(x_test_poly)

print(mean_squared_error(y_train, y_train_pred))
print(mean_squared_error(y_test, y_test_pred))
```

```python
ridge = Ridge(alpha=0.1)
ridge.fit(x_train_poly, y_train)

y_train_pred = ridge.predict(x_train_poly)
y_test_pred = ridge.predict(x_test_poly)

print(mean_squared_error(y_train, y_train_pred))
print(mean_squared_error(y_test, y_test_pred))
```
